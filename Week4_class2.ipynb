```python
input_number = input("Enter a number :")
if  input_number.replace('.',"").replace("-","").replace("+",'').isdigit():
  num = float(input_number)
  if num > 0:
      print("positive number")
  elif   num < 0:
      print("negative  Number")
  else:
    print("Number is zero")

else:
  print("Please write valid integer")

```
