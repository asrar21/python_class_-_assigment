```python
# dictionery
dict_1 = {"NAME":"ASRAR", "AGE":29,"SKILLS":["Data Science","AI","ML"]}
# dict keys
print(dict_1.keys())
#dict values()
print(dict_1.values())
```

```python
# dictionery item
print(dict_1.items())
```

```python
# dictionery get
print(dict_1.get("SKILLS"))
```

```python
# new dictionery person
person = {'Name':"Asrar",'Gender':"Male",'Marks':{'Python':70,'JS':75,'C#':60},'Age':29}
print(person)
```

```python
# get JS marks
print(person.get('Marks')['JS'])
```

```python
# add Dictioneries
print(len(dict_1))
print(len(person))
```

```python
# Concatenation of dictionery
person.update(dict_1)
print(person)
```

```python
# Take marks from updated Dict
print(person.get('Marks'))
```
