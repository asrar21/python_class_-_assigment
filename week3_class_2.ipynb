```python
# Add or Update in sets
s3 = {0,12,3,43}
s3.add(67)
s3.update([70,4,5,6,7,8,65,43]) # when to add multiple elements in set we have to use update method
print(s3)
```

```python
# Add or Update in set using tuple
s4 = {"a","b","c","d","e","f"}
s4.add("g")
s4.update(("imran"))
print(s4)
```

```python
# other case like this
name = {'imran','khan'}
print(name)
name.update({0,1,2,3,4}) # set is unordered
print(name)
```

```python
# remove element
name.remove(4)
print(name)
```

```python
# in set pop remove 1st element where as pop in other datatypes removes last element
name.pop()
print(name)
```

```python
# discard funtion
name.discard('imran')
print(name)
```

```python
# clear function clear the set
name.clear()
print(name)
```

```python
# union of set is (|)
s5 = {12,32,54,5,543}
s6 = {34,54,6,5765,4324}

print(s5 | s6)
```

```python
# intersection of set is (&)
s7 = {12,54,5}
s6 = {34,12,5}
print(s7 & s6)
```

```python
# difference of set is (-)
s8 = {3,2,4,5,6}
s9 = {3,4,5,10,12,13}
print(s8 - s9)
```

```python
# Symetric Differnce (^)
s10 = {3,2,4,5,6}
s11 = {3,4,5,10,12,13}
print(s10 ^ s11)
```

```python
# memebership
3 in s10
6 in s11
```

```python
# diffence in discard and remove
# discard only discard element which are present and not show error but remove show error when elements are not present
s10.discard(44)
# s10.remove(44)
s10.discard(44)
```

```python
# superset
s12 = {1,2,3,4,5}
s13 = {3,4,5}
s12.issuperset(s13)
```

```python
# disjoint completely different from set
s12 = {1,2,3,4,5}
s13 = {9,10,45}
s12.isdisjoint(s13)
```

```python
# subset
s12 = {1,2,3,4,5}
s13 = {3,4,5}
s13.issubset(s12)
```

```python
# Frozen set when we have to make our set immutable
# all read functions works
# all write function don't work
fs1 = frozenset([1,2,3])
fs = frozenset([1,2,frozenset([1,2,3])])
print(fs1)
print(fs)
```
