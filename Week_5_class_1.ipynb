```python
number_of_units = float(input("Enter your Units"))

if number_of_units <= 150:
  fc = 250
  pr_c = 15
elif   number_of_units > 150 and number_of_units < 250:
   fc = 350
   pr_c = 25
elif    number_of_units > 250 and number_of_units < 350:
  fc = 450
  pr_c = 35
else:
  fc = 1000
  pr_c = 45


print("your bill is : ",number_of_units * pr_c)
print("amount payable is :",(number_of_units * pr_c) + fc)
```

```python
#  Menu driven program
number1 = float(input("Enter Your First Number "))
number2 = float(input("Enter Your Second Number "))
option = input('''
1 - Addition
2 - Subtraction
3 - Multiplication
4 - Division
''')

if  option == '1':
  print("result is : ",number1 + number2)
elif option == '2':
    print("Result is : ",number1 - number2)
elif option == '3':
      print("Result is : ",number1 * number2)
elif option == '4':
        print("Result is : ",number1 / number2)
else:
  print("Wrong Menu Selected")
```
