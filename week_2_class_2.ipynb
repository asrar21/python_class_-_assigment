```python
# Decalrartion Syntax Of List
fruits = ["Mango","Banana","grapes","Avacado","Berry"];
print(type(fruits))
print("Declaration an Empty List",fruits)
print(len(fruits))
```

```python
# Adding Element in the List using append
fruits.append("Apple")
fruits.append("Dragon Fruits")
# when ever you add any value with append function it will be add into list
print(fruits)
```

```python
# indexing starts with 0
# print(fruits[0])
# print(fruits[0])
# print(fruits[])
print(fruits[:])# inclusive : Exclusive
# print(fruits[0:4])
# print(fruits[1:6])
print(fruits[0:7:1])# inclusive : Exclusive : Step 1 is by default impaact is done by 2
# skip 1 fruits
print(fruits[0:7:2])
# pick up grapes,berry
print(fruits[2:7:2])
# pick up banana, avacado
print(fruits[1:7:2])
# pick up banana, avacado
print(fruits[1:7:2])
```

```python
# Negative indexing [-3,-2,-1]
print(fruits[-1])
print(fruits[-5])
# 'grapes', 'Avacado', 'Berry', 'Apple'
print(fruits[-5:-1])
# banana avacado apple
print(fruits[-6:-1:2])
```
