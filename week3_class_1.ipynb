```python
# numbers tuple is imputable
tuple_of_number = (45,6,7,8,9,0,4,3,43,99,100,78,89,500)
tuple_of_number[0]
tuple_of_number[-1]
tuple_of_number.append()
```

```python
tuple_of_number.count(500)
```

```python
tuple_of_number1 = (7.9,9.0,8.9,2.3)
tuple_new_numbers= tuple_of_number + tuple_of_number1
print(tuple_new_numbers)
```

```python
tuple_new_numbers= tuple_of_number - tuple_of_number1
print(tuple_new_numbers)
```

```python
print(tuple_of_number *2)
# this concate tuple_of_number twice
# tuple will only support operator supported by string
```

```python
num1,num2,num3 = (1,2,3)# this is called unpacking or destructuring
print(type(num1))
print(num1,num2,num3)
```

```python
num1,*num3 = (1,2,3)# this is called unpacking or destructuring *num3 is others is the remaining
print(type(num3))
print(num1,num2)
print(num3)
```
