```python
df = pd.read_csv('./sample_data/googleplaystore.csv')
# df.head()
# df.tail()
# your primary duty to identify that columns are in correct format
# df.info()
# har specific columns ki misssing value bata daye ga
# df.isnull().sum()
# string filll with mode most repeated value
# mean aur median stdeviation and variance for numeric columns
# now price columns are in object we have to make it numberic
# df['Price'].unique()
# give all unique values you will identify the what values are not proper
# repalce 'Everyone' with zero and change datatype of columns
df['Price'] = df['Price'].replace('Everyone','0')
df['Price'] =  df['Price'].str.replace('$','').astype(float)
df['Price'].unique()
# to check datatype
# df.Price.dtype
df.info()
```

```python

# convert size colum to flaot
# df['Size'] = df['Size'].str.replace('M','').replace('k','').astype(float)
# df['Size'] = df['Size'].replace('Varies with device','0')
# df['Size'] = df['Size'].str.replace('k','')
df['Size'] = df['Size'].str.replace('+','')
df['Size'] = df['Size'].str.replace(',','')
df['Size']=df['Size'].astype(float)
df['Size'].unique()
# to remove duplicates
df.drop_duplicates(subset='App')
# df.shape gives roes and columns
df.shape()
```

```python
# to remove duplicates
df = df.drop_duplicates(subset='App')
# df.shape gives roes and columns
# df.shape
df
```

```python
# rating columns
df['Rating'].unique()
# how i dentify mean aur median calculae both if differnce bohat ziyada ho to mena se fill karaye gaye
df['Rating'].mean()
df['Rating'].median()
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
df['Rating'].unique()
```

```python
df['Reviews'] = df['Reviews'].str.replace('M','').astype(float)
df['Reviews'].unique()
```

```python
# df.head()
# now for numerilization of columns Reviews
# humesha data normalize hota hai between 0 and 1
# formula currentvalue - minvalue/maxvalue - minvalue
df['Reviews'] = (df['Reviews'] - df['Reviews'].min())/(df['Reviews'].max() - df['Reviews'].min())
df['Popularity'] = df['Reviews'] * df['Rating']
df.head()
```

```python
# to find which app is popular the most
df.sort_values(by='Popularity',ascending=False).head(3)
# df.head()
```
